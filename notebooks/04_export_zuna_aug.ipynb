# 04. Export FIF and run ZUNA augmentation

```
import os
import sys
from pathlib import Path

ROOT = Path.cwd() if (Path.cwd() / "configs").exists() else Path.cwd().parent
os.chdir(ROOT)
sys.path.insert(0, str(ROOT))

DATA_ROOT = ROOT / "data"
MANIFESTS = DATA_ROOT / "manifests"
RUNS = ROOT / "runs"

DATASET_ID = "ds005815"
TASK_CONTAINS = "task-task"
SKIP_ASR = True
SKIP_ICA = True

FIF_DIR = DATA_ROOT / "processed/fif_for_zuna"
ZUNA_WORK = RUNS / "zuna_yoto"

EPOCH_INDEX = MANIFESTS / "epoch_index_yoto_tones.csv"
ZUNA_INDEX = MANIFESTS / "epoch_index_yoto_tones_zuna.csv"
```

```
from scripts.build_zuna_augmented_index_yoto import build_zuna_epoch_index
from scripts.export_yoto_to_fif import export_yoto_fif
from scripts.run_zuna_augmentation import run_zuna_augmentation

export_summary = export_yoto_fif(
    dataset_id=DATASET_ID,
    task_contains=TASK_CONTAINS,
    out_dir=FIF_DIR,
    skip_asr=SKIP_ASR,
    skip_ica=SKIP_ICA,
)

zuna_summary = run_zuna_augmentation(
    input_fif_dir=FIF_DIR,
    work_dir=ZUNA_WORK,
)

zuna_index_summary = build_zuna_epoch_index(
    zuna_fif_dir=ZUNA_WORK / "4_fif_output",
    epoch_index=EPOCH_INDEX,
    out_dir=DATA_ROOT / "processed/epochs_aug",
    out_index=ZUNA_INDEX,
)

export_summary, zuna_summary, zuna_index_summary
```
