# 01. Download YOTO (and optional extras)

```
import os
import sys
from pathlib import Path

ROOT = Path.cwd() if (Path.cwd() / "configs").exists() else Path.cwd().parent
os.chdir(ROOT)
sys.path.insert(0, str(ROOT))

DATA_ROOT = ROOT / "data"
RAW_ROOT = DATA_ROOT / "raw_samples"
MANIFESTS = DATA_ROOT / "manifests"

YOTO_SUBJECTS = 5
EXTRA_DATASETS = []
EXTRA_MAX_SUBJECTS = 1
```

```
from scripts.download_sample_subjects import (
    download_openneuro_subjects,
    download_yoto_five,
    write_download_manifest,
)

downloads = [download_yoto_five(max_subjects=YOTO_SUBJECTS, raw_root=RAW_ROOT)]
if EXTRA_DATASETS:
    downloads.extend(
        download_openneuro_subjects(
            EXTRA_DATASETS,
            max_subjects=EXTRA_MAX_SUBJECTS,
            raw_root=RAW_ROOT,
        )
    )

write_download_manifest(downloads, manifest_path=MANIFESTS / "sample_download_manifest.json")
```
