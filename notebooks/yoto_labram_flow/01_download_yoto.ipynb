# 01 Download YOTO samples

Download YOTO (and optional extra datasets) and write a download manifest.

```
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
if not (ROOT / "scripts").exists():
    if (ROOT.parent / "scripts").exists():
        ROOT = ROOT.parent
    elif (ROOT.parent.parent / "scripts").exists():
        ROOT = ROOT.parent.parent

if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

DOWNLOAD_YOTO_FIVE = True
MAX_SUBJECTS = 5
SKIP_OTHER_DATASETS = True

# Add more OpenNeuro dataset IDs later if needed (e.g., ["ds004621"]).
EXTRA_OPENNEURO_DATASETS = []
INCLUDE_HAJONIDES = False
```

```
import json

from scripts.download_sample_subjects import (
    MANIFEST,
    RAW_ROOT,
    download_openneuro_subjects,
    download_yoto_five,
)

RAW_ROOT.mkdir(parents=True, exist_ok=True)

downloads = []
if DOWNLOAD_YOTO_FIVE:
    downloads.extend(
        download_yoto_five(
            max_subjects=MAX_SUBJECTS,
            skip_other=SKIP_OTHER_DATASETS,
        )
    )

if EXTRA_OPENNEURO_DATASETS:
    downloads.extend(
        download_openneuro_subjects(
            dataset_ids=EXTRA_OPENNEURO_DATASETS,
            include_hajonides=INCLUDE_HAJONIDES,
        )
    )
elif not DOWNLOAD_YOTO_FIVE:
    downloads.extend(download_openneuro_subjects(include_hajonides=INCLUDE_HAJONIDES))

MANIFEST.parent.mkdir(parents=True, exist_ok=True)
MANIFEST.write_text(json.dumps({"downloads": downloads}, indent=2), encoding="utf-8")

print(f"Wrote manifest: {MANIFEST}")
print(f"Downloads: {len(downloads)}")
```
