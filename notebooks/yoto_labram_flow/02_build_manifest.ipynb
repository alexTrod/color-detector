# 02 Build unified manifest

Scan downloaded files and build the unified manifest.

```
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
if not (ROOT / "scripts").exists():
    if (ROOT.parent / "scripts").exists():
        ROOT = ROOT.parent
    elif (ROOT.parent.parent / "scripts").exists():
        ROOT = ROOT.parent.parent

if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

RAW_ROOT = ROOT / "data/raw_samples"
OUT_JSON = ROOT / "data/manifests/unified_manifest.json"
OUT_CSV = ROOT / "data/manifests/unified_manifest.csv"
```

```
import pandas as pd

from scripts.build_unified_manifest import build_manifest, write_manifest

rows = build_manifest()
write_manifest(rows, OUT_JSON, OUT_CSV)

if rows:
    df = pd.DataFrame(rows)
    print(df.groupby("dataset_id").size().sort_values(ascending=False))
else:
    print("No EEG files found under", RAW_ROOT)
```
