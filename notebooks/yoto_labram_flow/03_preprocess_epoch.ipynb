# 03 Preprocess and epoch YOTO

Run YOTO preprocessing and generate tone epochs.

```
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
if not (ROOT / "scripts").exists():
    if (ROOT.parent / "scripts").exists():
        ROOT = ROOT.parent
    elif (ROOT.parent.parent / "scripts").exists():
        ROOT = ROOT.parent.parent

if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

MANIFEST = ROOT / "data/manifests/unified_manifest.csv"
OUT_DIR = ROOT / "data/processed/epochs_yoto_tones"
OUT_INDEX = ROOT / "data/manifests/epoch_index_yoto_tones.csv"

# YOTO datasets (add more dataset IDs later if needed).
DATASETS = ["ds005815"]
TASK_CONTAINS = "task-task"
SKIP_ASR = True
SKIP_ICA = True
```

```
from scripts.preprocess_yoto import run_preprocess_yoto

rows = run_preprocess_yoto(
    manifest=MANIFEST,
    datasets=DATASETS,
    task_contains=TASK_CONTAINS,
    out_dir=OUT_DIR,
    out_index=OUT_INDEX,
    skip_asr=SKIP_ASR,
    skip_ica=SKIP_ICA,
)

print(f"Epoch rows: {len(rows)}")
if rows:
    print("First row:", rows[0])
```
